```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1.0, 1.0, 100)
x
```

```python
#define parameters for quantization
min_x = -1.0
max_x = 1.0
qmin = -128
qmax = 127

#zero point for symetric quantization
scale = (max_x - min_x)/(qmax - qmin) 
scale
```

```python
zero_point = 0

#quantization to float -> int8
q = np.round((x-zero_point)/(scale)).astype(int)

q = np.clip(q,qmin,qmax)    
q
```

```python
#dequantization

deq = scale  * (q- zero_point)
deq
```

```python
#ploting quantized and dequantized

plt.figure(figsize=(10,5))
plt.plot(x,x,label = "Original Float (x)", linestyle = '--', color = 'gray')
plt.scatter(x,deq,label = 'Dequantized (X)', color = 'red', s =15)
plt.title('Linear Quantization int8 with Dequatization')
plt.xlabel('Original Float Value')
plt.ylabel('Dequantized Value')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# define 16 NF4 representative float values
nf4_values = np.array([
    -1.0, -0.75, -0.5, -0.35, -0.2, -0.1, -0.05, -0.01,
     0.01,  0.05,  0.1,  0.2,  0.35,  0.5,  0.75,  1.0
])

x = np.linspace(-1.0,1.0,100)
x
```

```python
def quantize_nf4(x_vals, nf4_lut):
    indices = np.argmin(np.abs(x_vals[:,None] - nf4_lut[None, :]), axis=1)
    return nf4_lut[indices]

x_nf4 = quantize_nf4(x,nf4_values)
x_nf4
```

```python
#ploting original and quantized
plt.figure(figsize=(10,5))
plt.plot(x,x,label = 'Original Float x', linestyle='--', color = 'gray')
plt.scatter(x,x_nf4,label='Quantized to NF4' , color='blue', s=15)
plt.title("NF4 Quantization using 16 Discrete values")
plt.xlabel('Original Float Values')
plt.ylabel('Quantized NF4 Values')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
